```python
cuboid = mesh.Mesh.from_file("/content/Vase 1.stl")

x_min, y_min, z_min = cuboid.min_
x_max, y_max, z_max = cuboid.max_
width = int(x_max) - int(x_min)
height = int(y_max) - int(y_min)
depth = int(z_max) - int(z_min)

print(x_min, y_min, z_min)
print(x_max, y_max, z_max)

MASK = np.zeros((width+1, height+1, depth+1))

print(MASK.shape)
input("PRESS ENTER TO CONTINUE...")

for triangle in cuboid.vectors:
    vertices = np.array(triangle)
    x = vertices[:, 0]
    y = vertices[:, 1]
    z = vertices[:, 2]
    x = [i-x_min for i in x]
    y = [i-y_min for i in y]
    z = [i-z_min for i in z]
    for i in range(int(min(x)), int(max(x)+1)):
      for j in range(int(min(y)), int(max(y)+1)):
        for k in range(int(min(z)), int(max(z)+1)):
          p = np.array([i, j, k])
          p1 = np.array(vertices[0])
          p2 = np.array(vertices[1])
          p3 = np.array(vertices[2])

          # Compute triangle normal
          v0 = p3 - p1
          v1 = p2 - p1
          n = np.cross(v0, v1)

          # Check if point is on the same side of each triangle edge
          not_in = 1
          for l in range(3):
              edge_start = p1 if l == 0 else (p2 if l == 1 else p3)
              edge_end = p2 if l == 0 else (p3 if l == 1 else p1)
              edge = edge_end - edge_start
              vp = p - edge_start
              c = np.cross(edge, vp)
              if np.dot(n, c) < 0:
                  not_in = 0
                  break
          MASK[i][j][k] = not_in

plt.imshow(MASK[:, :, 50])
plt.plot()
```

```python
import numpy as np
from stl import mesh

def stl_to_3d_mask(stl_file_path, resolution):
    # Load STL file
    stl_mesh = mesh.Mesh.from_file(stl_file_path)

    # Determine dimensions of the mask
    min_coords = np.min(stl_mesh.points, axis=0)
    max_coords = np.max(stl_mesh.points, axis=0)
    dimensions = np.ceil((max_coords - min_coords) / resolution).astype(int)

    # Create an empty 3D mask
    mask = np.zeros(dimensions, dtype=bool)

    # Iterate through each face of the mesh
    for face in stl_mesh.vectors:
        # Calculate the face's bounding box
        min_face = np.floor(np.min(face, axis=0) / resolution).astype(int)
        max_face = np.ceil(np.max(face, axis=0) / resolution).astype(int)

        # Set the corresponding region in the mask to True
        mask[min_face[0]:max_face[0]+1, min_face[1]:max_face[1]+1, min_face[2]:max_face[2]+1] = True

    return mask

# Example usage
stl_file_path = "path/to/your/file.stl"
resolution = 0.1  # Adjust the resolution according to your requirements
mask = stl_to_3d_mask(stl_file_path, resolution)
```
